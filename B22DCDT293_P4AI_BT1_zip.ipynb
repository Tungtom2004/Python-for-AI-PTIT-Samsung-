**4.Chuẩn hóa dữ liệu theo kiểu Min Max**

```python
df4 = df1.copy()
```

```python
sepal_length_min = df4['sepal.length'].min()
sepal_length_max = df4['sepal.length'].max()

sepal_width_min = df4['sepal.width'].min()
sepal_width_max = df4['sepal.width'].max()

petal_length_min = df4['petal.length'].min()
petal_length_max = df4['petal.length'].max()

petal_width_min = df4['petal.width'].min()
petal_width_max = df4['petal.width'].max()
```

```python
#Thực hiện chuẩn hóa min max với 4 trường dữ liệu đã cho
df4['sepal.length'] = (df4['sepal.length'] - sepal_length_min )/(sepal_length_max - sepal_length_min)
df4['sepal.width'] = (df4['sepal.width'] - sepal_width_min)/(sepal_width_max - sepal_width_min)
df4['petal.length'] = (df4['petal.length'] - petal_length_min)/(petal_length_max - petal_length_min)
df4['petal.width'] = (df4['petal.width'] - petal_width_min)/(petal_width_max - petal_width_min)
```

**9. Phân cụm dữ liệu và loại bỏ ngoại lai**

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from scipy.spatial.distance import cdist
scaler = StandardScaler()
df5_scaled = scaler.fit_transform(df5[['sepal.length', 'sepal.width']])
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(df5_scaled)
df5['cluster'] = kmeans.labels_
distances = cdist(df5_scaled, kmeans.cluster_centers_, 'euclidean')
df5['distance_to_centroid'] = np.min(distances, axis=1)
threshold = df5['distance_to_centroid'].quantile(0.95)
df_no_outliers = df5[df5['distance_to_centroid'] <= threshold]
print(df_no_outliers.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.scatterplot(x=df5['sepal.length'], y=df5['sepal.width'], hue=df5['cluster'], style=df5['distance_to_centroid'] > threshold, palette='Set1', markers=['o', 'X'])
plt.title('Phân cụm dữ liệu và loại bỏ ngoại lai')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(title='Cluster')
plt.show()
```
